```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
from datetime import datetime

# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

today = datetime.today().strftime('%m-%d-%Y')
print(today)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# Call to api to get data
x = 1

city_list = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

for city in cities:
    try:
        base_url = "http://api.openweathermap.org/data/2.5/weather?q={0}&appid={1}".format(city, api_key)
        print("Processing record " + str(x) + " of set 1 | " + city)
        response = requests.get(base_url).json()

        x = x+1
        #pprint(response)
        
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        max_temp.append(response["main"]["temp_max"])
        wind_speed.append(response["wind"]["speed"])
        city_list.append(city)
        time.sleep(.75)
    except Exception:
        print("City not found. Skipping...")
        time.sleep(.75)
```

```python
df = pd.DataFrame({"City": city_list,
                   "Cloudiness": cloudiness,
                   "Country": country,
                   "Date": date,
                   "Humidity": humidity,
                   "lat": lat,
                   "lng": lng,
                   "Max Temp": max_temp,
                   "Wind Speed": wind_speed})
df.to_csv("weather.csv")
```

```python
df.plot.scatter(x = "lat",
               y = "Max Temp",
               c = "DarkBlue")
plt.title("Temperature vs Latitude (" + str(today) + ")")
plt.savefig("TempvsLat.png")
plt.show()
```

```python
df.plot.scatter(x = "lat",
               y = "Humidity",
               c = "DarkBlue")
plt.title("Humidity vs Latitude (" + str(today) + ")")
plt.savefig("HumvsLat.png")
plt.show()
```

```python
df.plot.scatter(x = "lat",
               y = "Cloudiness",
               c = "DarkBlue")
plt.title("Cloudiness vs Latitude (" + str(today) + ")")
plt.savefig("CloudvsLat.png")
plt.show()
```

```python
df.plot.scatter(x = "lat",
               y = "Wind Speed",
               c = "DarkBlue")
plt.title("Wind Speed vs Latitude (" + str(today) + ")")
plt.savefig("WindvsLat.png")
plt.show()
```

Observation 1: Temperature is correlated with latitude. As the latitude approaches 0 (the equator) we see the temperatur at its' highest point
